# Recência
Avalia o período de tempo da última compra realizada até a última compra realizada no sistema

```python
## criando uma nova base para recencia
df_recencia = (
    base_geral[['cpf', 'data_criacao']]
    .groupby('cpf')['data_criacao']
    .agg('max')
    .reset_index()
)

# primeiras linhas
df_recencia.head()
```

```python
# usando a data mais recente
cur_time = df_recencia.data_criacao.max()
cur_time
```

```python
# calculo da diferença entre as datas em meses
def mes_diff(cur_time, datetime_val):
    return 12 * (cur_time.year - datetime_val.year) + (
        cur_time.month - datetime_val.month
    )

# criando uma nova coluna de recencia
df_recencia['recencia'] = df_recencia.data_criacao.apply(
    lambda data_criacao: mes_diff(cur_time, data_criacao)
)

# head
df_recencia.head()
```

```python
# plotagem do histograma
plt.hist(df_recencia['recencia'], bins = 60)
plt.title('Histograma de recencia')
plt.xlabel('Recência')
plt.ylabel('Frequência')
plt.show()
```

```python
# criando uma base unica
df_base_temp = df_base_temp.merge(df_cust_temp, on = 'cpf', how = 'left')
df_base_temp = df_base_temp.fillna(0)
df_base_temp.head()
```

```python
df_base_temp.hist(['QTDE_CUST'], bins = 60)
plt.show()
```

```python
## Cálculo para saber se a compra PF foi anterior à compra PJ
# conversão das colunas de tempo e tratamento dos valores zerados

df_base_temp['primeira_compra'] = pd.to_datetime(df_base_temp['primeira_compra'], errors = 'coerce')
df_base_temp['PRIMEIRA_COMPRA'] = pd.to_datetime(df_base_temp['PRIMEIRA_COMPRA'], errors = 'coerce')

# Função para calcular a diferença entre as datas
def calcular_diferenca(row):
    if pd.isna(row['primeira_compra']) or pd.isna(row['PRIMEIRA_COMPRA']) or row['QTDE_CUST'] == 0:
        return pd.NA
    else:
        return row['primeira_compra'] - row['PRIMEIRA_COMPRA']

# Aplicar a função ao DataFrame
df_base_temp['diferenca'] = df_base_temp.apply(calcular_diferenca, axis=1)
```

```python
# Converter para datetime usando errors='coerce' irá converter strings inválidas para NaT
df_base_temp['primeira_compra'] = pd.to_datetime(df_base_temp['primeira_compra'], errors='coerce')
df_base_temp['PRIMEIRA_COMPRA'] = pd.to_datetime(df_base_temp['PRIMEIRA_COMPRA'], errors='coerce')

# Função para calcular a diferença entre as datas
def calcular_diferenca(row):
    if pd.isna(row['primeira_compra']) or pd.isna(row['PRIMEIRA_COMPRA']) or row['QTDE_CUST'] == 0:
        return pd.NA
    else:
        # Certifique-se de que a operação de subtração ocorrerá apenas se ambas as datas forem válidas
        return (row['primeira_compra'] - row['PRIMEIRA_COMPRA']).days

# Aplicar a função ao DataFrame
df_base_temp['diferenca'] = df_base_temp.apply(calcular_diferenca, axis=1)

# Agora você pode converter a coluna 'diferenca' diretamente para int, se necessário, ou deixar como está se quiser manter o pd.NA
# df_base_temp['diferenca'] = df_base_temp['diferenca'].astype('Int64')  # Use 'Int64' (capital "I") para permitir pd.NA

# Verifique o DataFrame resultante
print(df_base_temp)
```
